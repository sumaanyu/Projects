#### Q 4.
The data files needed: 

`google_data.txt, ny_temps.txt & yahoo_data.txt`

Use your knowledge with `Python, NumPy, pandas and matplotlib` to reproduce the plot below:

![](answer.png?raw=true)

```python
google = pd.read_csv('google_data.txt', sep='\t') 
yahoo = pd.read_csv('yahoo_data.txt', sep='\t') 
ny = pd.read_csv('ny_temps.txt', sep='\t') 
```

```python
display(yahoo.head(), google.head(), ny.head())
```

```python
fig = plt.figure(figsize = [12,8])
matplotlib.rc('axes',edgecolor='black')
ax1 = fig.add_subplot(111)
ax1.plot('Modified Julian Date', 'Stock Value', data = yahoo, color = 'purple')
ax1.plot('Modified Julian Date', 'Stock Value', data = google, color = 'green')
ax1.set_ylabel('Value (Dollars)', color='purple')
ax1.set_xlabel("Date (MJD)")
for t0 in ax1.get_yticklabels():
    t0.set_color('green')
ax1.grid(False)
ax1.set_facecolor('white')
ax2 = ax1.twinx() 
ax2.grid(False)
ax2.plot("Modified Julian Date", 'Max Temperature', data = ny, color = 'blue', linestyle='--')
ax2.set_ylabel('Temperature (º F)', color='blue')
plt.ylim(-150, 100)
for tl in ax2.get_yticklabels():
    tl.set_color('blue')
ax2.legend(loc='center left', frameon=False)
ax1.legend(loc='center left', frameon=False);
ax1.set_title('New York Temperature, Google and Yahoo!', fontsize = 22) ;
```
